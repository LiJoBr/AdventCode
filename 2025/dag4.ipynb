```python
import pandas as pd
```

```python
bestand = r'.\data\4.txt'
df_rolls = pd.read_csv(bestand, header=None)
df_rolls = (df_rolls[0]
            .apply(list)
            .apply(pd.Series)
            .replace({'.': 0, '@': 1})
)

# add zero grid around
df_rolls.loc[-1] = [0] * df_rolls.shape[1]
df_rolls.loc[df_rolls.shape[0]-1] = [0] * df_rolls.shape[1]   
df_rolls.index = df_rolls.index + 1  
df_rolls = df_rolls.sort_index()
df_rolls.columns = df_rolls.columns + 1
df_rolls.insert(loc=0,column=0, value=[0] * df_rolls.shape[0])
df_rolls.insert(loc=df_rolls.shape[1],column=df_rolls.shape[1], value=[0] * df_rolls.shape[0])
```

# Part 1

```python
n_rows = df_rolls.shape[0] - 2
n_cols = df_rolls.shape[1] - 2
sum_acces = 0
for row in range(1, n_rows+1):
    for col in range(1, n_cols+1):
        sum = int(
            df_rolls.iloc[row-1:row+2,col-1:col+2]
            .sum()
            .sum()
            - df_rolls.iloc[row, col]
            )
        if (int(df_rolls.iloc[row, col]) == 1) & (sum < 4):
            sum_acces += 1
print(sum_acces)
```

# Part 2

```python

def remove_count(df, positions):
    sum_acces = 0
    for pos in positions:
        row = pos[0]
        col = pos[1]

        sum = int(
            df.iloc[row-1:row+2,col-1:col+2]
            .sum()
            .sum()
            - df.iloc[row, col]
            )
        if (sum < 4):
            sum_acces += 1
            df.iloc[row, col] = 0
            positions.remove(pos)
    if sum_acces > 0:
        sum_acces += remove_count(df, positions)
    return sum_acces
```

```python
df = df_rolls.copy()
positions = (df == 1).stack()
positions = positions[positions].index.tolist()
remove_count(df, positions)
```

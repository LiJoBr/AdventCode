```python
bestand = r'.\data\5.txt'
with open(bestand, 'r') as file:
    input = (file
            .read()
            .split("\n\n")      
            )
ranges = input[0].splitlines()
ids = input[1].splitlines()
ids = [int(id) for id in ids]
ranges = [tuple(map(int, item.split('-'))) for item in ranges]
ranges = sorted(ranges, key=lambda x: x[0])
```

```python
def in_any_range(x, ranges):
    return any(a <= x <= b for a, b in ranges)
```

```python
som = 0
for id in ids:
    som += in_any_range(id, ranges)
print(som)
```

```python
def combine(set1, set2):
    if set1[0] > set2[0]:
        s1 = set1
        set1 = set2
        set2 = s1
    if set1[0] == set2[0]:
        return [(set1[0], max(set1[1], set2[1]))]
    # set1[0] < set2[0]
    if set1[1] < set2[0]:
        return [set1, set2]
    if set1[1] == set2[0]:
        return [(set1[0], set2[1])]
    # set1[1] > set2[0]
    if set1[1] < set2[1]:
        return [(set1[0], set2[1])]
    else:
        return [(set1[0], set1[1])]   
            
```

```python
assert combine((1,2), (3,4)) == [(1,2), (3,4)]
assert combine((1,2), (6,7)) == [(1,2), (6,7)]
assert combine((1,6), (5,7)) == [(1,7)]
assert combine((1,6), (5,5)) == [(1,6)]
assert combine((1,6), (4,5)) == [(1,6)]
assert combine((1000, 1010), (1001, 1011)) == [(1000,1011)]
```

```python
def make_unique_ranges(ranges):
    if len(ranges) == 1:
        return ranges
    unique_range = []
    last = ranges[0]
    for i in range(1, len(ranges)):
        new = combine(last, ranges[i])
        if len(new) == 2:
            if i == len(ranges)-1:
                unique_range += new
            else:
                unique_range += [new[0]]
                last = new[1]
        else:
            last = new[0]
            if i == len(ranges)-1:
                unique_range += new
    return unique_range
```

```python
assert make_unique_ranges([(1,2), (3,4)]) == [(1,2), (3,4)]
assert make_unique_ranges([(1,7), (3,4), (4,5)]) == [(1,7)]
assert make_unique_ranges([(1,7), (3,4), (10,12)]) == [(1,7), (10,12)]
assert make_unique_ranges([(1,7), (7,7), (10,12)]) == [(1,7), (10,12)]
assert make_unique_ranges([(1,7), (9,9), (11,12)]) == [(1,7), (9,9), (11,12)]
assert make_unique_ranges([(1000, 1010), (1001, 1011)]) == [(1000,1011)]
assert make_unique_ranges([(1,8)]) == [(1,8)]
```

```python

def count_fresh_ids(ranges):
    count = 0
    ranges = make_unique_ranges(ranges)
    for range in ranges:
        count += range[1] - range[0] + 1
    return count
            
```

```python
assert count_fresh_ids([(1000, 1010), (1001, 1011)]) == 12
assert count_fresh_ids([(1, 8), (4, 5)]) == 8
assert count_fresh_ids([(1, 8), (8, 8)]) == 8
assert count_fresh_ids([(1, 8), (8, 9)]) == 9
assert count_fresh_ids([(1, 8)]) == 8
assert count_fresh_ids([(1, 8), (1,7)]) == 8
assert count_fresh_ids([(1, 8), (1,9)]) == 9
assert count_fresh_ids([(1, 8), (2,7)]) == 8
assert count_fresh_ids([(1, 8), (2,9)]) == 9
assert count_fresh_ids([(1, 8), (8,9)]) == 9
assert count_fresh_ids([(1, 8), (8,8)]) == 8
assert count_fresh_ids([(1, 8), (9,10)]) == 10
assert count_fresh_ids([(3, 5), (10, 14)]) == 8
assert count_fresh_ids([(3, 5), (10, 14), (12, 18)]) == 12
assert count_fresh_ids([(3, 5), (10, 14), (12, 18), (16, 20)]) == 14
```

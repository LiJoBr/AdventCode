# Deel 1

```python
bestand = r'.\data\1a.txt'

with open(bestand, 'r') as file:
    instructies = (file
            .read()
            .splitlines()
            )

```

```python
def calculate_next_place(place, instructie):
    """next place given instruction"""
    if instructie[0] == 'L':
        direction = -1
    else: 
        direction = 1

    move_places = int(instructie[1:]) % 100
    next_place = place + move_places * direction

    if (next_place < 0):
        next_place = 100 + next_place 
    
    if (next_place > 99):
        next_place = next_place - 100 

    return next_place
```

```python
def calculate_times_at_zero(instructies: list, start_place: int = 50):
    """Calculate times at zero"""
    place = start_place
    times_at_zero = 0

    for instructie in instructies:
        # print(place)
        place = calculate_next_place(place, instructie)
        if place == 0:
            times_at_zero += 1
        # print(instructie)
        # print(place)
        # print('-----')
    return times_at_zero
```

```python
assert calculate_next_place(0, 'R48') == 48
assert calculate_next_place(0, 'L2') == 98
assert calculate_next_place(1, 'L2') == 99
assert calculate_next_place(99, 'R2') == 1
assert calculate_next_place(99, 'R102') == 1
assert calculate_next_place(1, 'L102') == 99
assert calculate_next_place(1, 'L202') == 99
assert calculate_next_place(99, 'R202') == 1
assert calculate_next_place(60, 'R207') == 67
```

```python
calculate_times_at_zero(instructies)
```

```python
def calculate_next_place_and_count_zeros_hits(place, instructie):
    """next place given instruction"""
    if instructie[0] == 'L':
        direction = -1
    else: 
        direction = 1

    zero_hits = 0
    move_places = int(instructie[1:]) % 100
    next_place = place + move_places * direction

    # print('startplace', place)
    # print('instructie', instructie)
    # print('next_place without correctie', next_place)
   
    if (next_place < 0):
        next_place = 100 + next_place 
        if place != 0:
            zero_hits += 1
    
    if (next_place > 99):
        next_place = next_place - 100
        zero_hits += 1

    # print('zero_hits1', zero_hits)

    multiple_hunderd = math.floor(int(instructie[1:])/100)
    zero_hits += multiple_hunderd
    if (next_place == 0) & (zero_hits > 0):
        zero_hits -= 1
    # print('multiple_hunderd', multiple_hunderd)
    # print('final zero hits', zero_hits)
    # print('final next place', next_place)
    # print('-------')
    return next_place, zero_hits
```

```python
assert calculate_next_place_and_count_zeros_hits(0, 'R48') == (48, 0)
assert calculate_next_place_and_count_zeros_hits(0, 'L2') == (98, 0)
assert calculate_next_place_and_count_zeros_hits(1, 'L2') == (99,1)
assert calculate_next_place_and_count_zeros_hits(99, 'R2') == (1,1)
assert calculate_next_place_and_count_zeros_hits(99, 'R102') == (1,2)
assert calculate_next_place_and_count_zeros_hits(1, 'L102') == (99,2)
assert calculate_next_place_and_count_zeros_hits(99, 'R202') == (1,3)
assert calculate_next_place_and_count_zeros_hits(60, 'R207') == (67,2)
assert calculate_next_place_and_count_zeros_hits(52, 'R48') == (0,0)
assert calculate_next_place_and_count_zeros_hits(52, 'L52') == (0,0)
assert calculate_next_place_and_count_zeros_hits(56, 'L56') == (0,0)
assert calculate_next_place_and_count_zeros_hits(49, 'R151') == (0,1)
assert calculate_next_place_and_count_zeros_hits(50, 'R1000') == (50,10)
```

```python
def calculate_times_at_zero(instructies: list, start_place: int = 50):
    """Calculate times at zero"""
    place = start_place
    times_at_zero = 0
    times_through_zero = 0

    for instructie in instructies:
        print('startplaceplace', place)
        (place, times_through_zero_now) = calculate_next_place_and_count_zeros_hits(place, instructie)
        if place == 0:
            times_at_zero += 1
        times_through_zero += times_through_zero_now
        print('instructie', instructie)
        print('finish place', place)
        print('times through zero this time', times_through_zero_now)
        print('times_total_through_zero',times_through_zero )
        print('-----')

    return times_at_zero, times_at_zero+times_through_zero  
```

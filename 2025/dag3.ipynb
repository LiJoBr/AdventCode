```python
bestand = r'.\data\3a.txt'

with open(bestand, 'r') as file:
    joltage = (file
            .read()
            .splitlines()         
            )

```

```python
def switch(j, n):
    i = -1
    d = ''
    for k in range(n, 0, -1): 
        if k-1 == 0:
            s = max(int(x) for x in j[i+1:])
            i = len(j[0:i+1]) + j[i+1:].index(str(s))
        else:
            s = max(int(x) for x in j[i+1:-(k-1)])
            i = len(j[0:i+1]) + j[i+1:-(k-1)].index(str(s))
        d += str(s)
    return int(d)
```

```python
som2 = 0
som12 = 0
for j in joltage:
    som2 += switch(j, 2)
    som12 += switch(j, 12)
print(som2)
print(som12)
```

```python
bestand = r'.\data\2a.txt'

with open(bestand, 'r') as file:
    ids = (file
            .read()          
            .split(',')
            )

```

```python
def invalid_1(x):
    number_digits = len(str(x))   
    if number_digits % 2 != 0:
        return False
    half_double_number = str(x)[:int(number_digits/2)]
    double_number = int(half_double_number + half_double_number)
    if x == double_number:
        return True
    else:
        return False
    
```

```python
def invalid_2(x):
    s = str(x)
    prefixes = [s[:i] for i in range(1, len(s)//2+1)]
    for prefix in prefixes:
        if len(s) % len(prefix) == 0:
            r = [prefix for i in range(1, len(s) // len(prefix)+1)]
            if int("".join(r)) == x:
                return True
    return False
```

```python
som_1 = 0
som_2 = 0
for id in ids:
    id = id.split('-')
    start = int(id[0])
    end = int(id[1])
    for x in range(start, end+1):
        if invalid_1(x):
            som_1 += x
        if invalid_2(x):
            som_2 += x
            
print(som_1)
print(som_2)
```
